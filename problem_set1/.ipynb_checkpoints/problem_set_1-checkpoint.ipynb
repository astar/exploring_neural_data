```python
spikes = np.load('spikes_hard_practice_answers.npy')[()]
data = np.load('spikes_hard_practice.npy')[()]
time, voltage = np.array(data['time']), np.array(data['voltage'])
```

```python
for spike in spikes:
    print voltage[time==spike]
```

```python
APTime = spikes[2]        
APRange = 3e-3
start = APTime - APRange 
stop = APTime + APRange

y = voltage[(time > start) & (time < stop)]
x = np.linspace(-APRange, APRange, len(y))
plt.plot(x, y)
plt.axvline(x=x[y==max(y)]+.0002, linewidth=1, color='red')
```

```python
time[voltage == max(voltage[(time > p) & (time < p + JITTER)])][0]
```

```python
idxs = (voltage - np.mean(voltage))/np.std(voltage) > 5
potencials = time[idxs]
JITTER = 0.001
tmp = []
for p in potencials:
    local_idxs = (time > p - JITTER) & (time < p + JITTER)
    local_time = time[local_idxs]
    local_voltage = voltage[local_idxs]
    tmp.append(local_time[local_voltage == max(local_voltage)][0] + 0.00024)
my_spikes =  np.unique(tmp)
my_spikes
```
